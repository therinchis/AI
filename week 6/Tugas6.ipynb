```python
import pandas as pd
from sklearn.naive_bayes import CategoricalNB
from sklearn.preprocessing import OrdinalEncoder
```

```python
#data tugas
data = {
    'Harga': ['Murah', 'Murah', 'Sedang', 'Mahal', 'Mahal', 'Sedang', 'Sedang', 'Mahal'],
    'Diskon': ['Ada', 'Ada', 'Ada', 'Tidak Ada', 'Tidak Ada', 'Ada', 'Tidak Ada', 'Ada'],
    'Ulasan': ['Baik', 'Baik', 'Baik', 'Buruk', 'Buruk', 'Baik', 'Baik', 'Buruk'],
    'Hasil': ['Tertarik', 'Tertarik', 'Tertarik', 'Tidak Tertarik', 'Tidak Tertarik', 'Tertarik', 'Tertarik', 'Tidak Tertarik']
}

df = pd.DataFrame(data)
print("Data Latih:")
display(df)
```

Pra-pemrosessan Data (Encoding)

```python
# Memisahkan fitur (X) dan target (y)
X = df[['Harga', 'Diskon', 'Ulasan']]
y = df['Hasil']

# Inisialisasi encoder
encoder = OrdinalEncoder()

# Fit dan transform data fitur menjadi numerik
X_encoded = encoder.fit_transform(X)

print("\nRepresentasi Numerik dari Fitur (X):")
print(X_encoded)
```

Model Training

```python
# Inisialisasi model Naive Bayes Kategorikal dengan Laplace Smoothing
# alpha=1.0 akan menambahkan 1 ke setiap frekuensi fitur untuk mencegah probabilitas 0
model = CategoricalNB(alpha=1.0)

# Melatih model dengan data numerik
model.fit(X_encoded, y)

print("Model Categorical Naive Bayes berhasil dilatih.")
```

Prediksi Data Baru

```python
# Mendefinisikan data baru
new_data = pd.DataFrame([['Sedang', 'Ada', 'Baik']], columns=['Harga', 'Diskon', 'Ulasan'])

new_data_encoded = encoder.transform(new_data)

# Memprediksi kelas (Hasil)
prediksi_kelas = model.predict(new_data_encoded)

# Melihat probabilitas dari masing-masing kelas
probabilitas = model.predict_proba(new_data_encoded)

print(f"Data Pelanggan Baru: Harga={new_data.iloc[0]['Harga']}, Diskon={new_data.iloc[0]['Diskon']}, Ulasan={new_data.iloc[0]['Ulasan']}")
print(f"\n>> PREDIKSI HASIL: {prediksi_kelas[0]} <<")

# Menampilkan probabilitas di balik layar
kelas_target = model.classes_
print("\nProbabilitas Posterior (Setelah Smoothing dan Normalisasi):")
for i, kelas in enumerate(kelas_target):
    # Mengalikan dengan 100 untuk format persentase
    print(f"- {kelas}: {probabilitas[0][i] * 100:.2f}%")
```
